```python
import os
import pandas as pd
```

```python
# Укажите путь к папке
folder_path = 'pzd_1'
# Получаем список файлов в папке
files = os.listdir(folder_path)
# Берем первый файл из списка
first_file = files[0]
print(first_file)
```

```python
# Формируем полный путь к файлу
file_path = os.path.join(folder_path, first_file)
# Читаем CSV-файл с помощью Pandas
df = pd.read_csv(file_path)
```

```python
pd.set_option('display.max_columns', None)
```

```python
df.head()
```

```python
df.tail()
```

```python
import matplotlib.pyplot as plt
date = first_file.split('.')[0]
plt.plot(df['days_to_otp'],df['target_count_tickets_lgot_255'])
plt.xticks(ticks=range(min(df['days_to_otp']), max(df['days_to_otp']) + 1, 5))
plt.title(f'Дата отправления {date}')
plt.grid()
plt.xlabel('Дней до отправления')
plt.ylabel('Кол-во проданных билетов на каждый день')
plt.tight_layout()
plt.show()
```

```python
plt.plot(df['days_to_otp'],df['delta_tickets_lgot_255'])
plt.xticks(ticks=range(min(df['days_to_otp']), max(df['days_to_otp']) + 1, 5))
plt.title(f'Дата отправления {date}')
plt.xlabel('Дней до отправления')
plt.ylabel('Кол-во проданных билетов на дату отправления')
plt.gca().invert_xaxis()
plt.grid()
plt.tight_layout()
plt.show()
```
